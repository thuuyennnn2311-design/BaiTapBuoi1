```python
#14
import random

# a. Tạo mảng ngẫu nhiên
def tao_mang(m, n):
    a = []
    for i in range(m):
        dong = []
        for j in range(n):
            dong.append(random.randint(0, 99))
        a.append(dong)
    return a

# Xuất mảng
def xuat_mang(a):
    for dong in a:
        for x in dong:
            print(f"{x:4}", end="")
        print()

# b. Xuất dòng k
def xuat_dong_k(a, k):
    print("Dòng", k, ":", a[k])

# c. Xuất cột k
def xuat_cot_k(a, k):
    print("Cột", k, ":")
    for dong in a:
        print(dong[k])

# d. Tìm dòng có tổng > 45 lớn nhất
def tim_dong(a):
    max_tong = -1
    vi_tri = -1
    for i in range(len(a)):
        tong = sum(a[i])
        if tong > 45 and tong > max_tong:
            max_tong = tong
            vi_tri = i
    return vi_tri

# e. Tìm cột có tích nhỏ nhất
def tim_cot(a):
    m = len(a)
    n = len(a[0])
    min_tich = None
    vi_tri = -1

    for j in range(n):
        tich = 1
        for i in range(m):
            tich *= a[i][j]

        if min_tich is None or tich < min_tich:
            min_tich = tich
            vi_tri = j

    return vi_tri

# f. Xuất phần tử dòng chẵn, cột lẻ
def xuat_chan_le(a):
    print("Phần tử dòng chẵn, cột lẻ:")
    for i in range(len(a)):
        for j in range(len(a[0])):
            if i % 2 == 0 and j % 2 != 0:
                print(a[i][j], end=" ")
    print()

# g. Trung bình cộng số chẵn thuộc dòng lẻ
def tbc_chan_dong_le(a):
    tong = 0
    dem = 0
    for i in range(len(a)):
        if i % 2 != 0:
            for j in range(len(a[0])):
                if a[i][j] % 2 == 0:
                    tong += a[i][j]
                    dem += 1
    return tong / dem if dem != 0 else 0

# h. Trung bình cộng phần tử biên
def tbc_bien(a):
    m = len(a)
    n = len(a[0])
    tong = 0
    dem = 0

    for i in range(m):
        for j in range(n):
            if i == 0 or i == m-1 or j == 0 or j == n-1:
                tong += a[i][j]
                dem += 1
    return tong / dem
#i. TBC không thuộc biên
def tbc_khong_bien(a):
    m = len(a)
    n = len(a[0])

    tong = 0
    dem = 0

    for i in range(1, m-1):
        for j in range(1, n-1):
            tong += a[i][j]
            dem += 1

    if dem == 0:
        return 0
    return tong / dem

# ================== MAIN ==================
m = int(input("Nhập số dòng: "))
n = int(input("Nhập số cột: "))

a = tao_mang(m, n)

print("\nMa trận:")
xuat_mang(a)

k = int(input("\nNhập k: "))
xuat_dong_k(a, k)
xuat_cot_k(a, k)

print("\nDòng có tổng > 45 lớn nhất:", tim_dong(a))
print("Cột có tích nhỏ nhất:", tim_cot(a))

xuat_chan_le(a)

print("Trung bình cộng số chẵn dòng lẻ:", tbc_chan_dong_le(a))
print("Trung bình cộng biên:", tbc_bien(a))
print("Trung bình cộng các phần tử không thuộc biên:",tbc_khong_bien(a))
```

```python
#15
class SinhVien:
    def __init__(self, ma, ten, nam_sinh, diem):
        self.ma = ma
        self.ten = ten
        self.nam_sinh = nam_sinh
        self.diem = diem

#a
def nhap_danh_sach():
    ds = []
    n = int(input("Nhập số sinh viên: "))
    for i in range(n):
        print(f"\nSinh viên {i+1}:")
        ma = input("Mã SV: ")
        ten = input("Tên: ")
        nam_sinh = int(input("Năm sinh: "))
        diem = float(input("Điểm TB: "))
        ds.append(SinhVien(ma, ten, nam_sinh, diem))
    return ds

#c. Đếm SV lên lớp (điểm >=5)
def dem_len_lop(ds):
    dem = 0
    for sv in ds:
        if sv.diem >= 5:
            dem += 1
    return dem
#d. Xuất SV đủ 20 tuổi ===================
def sv_20_tuoi(ds):
    print("\nSinh viên đủ 20 tuổi:")
    for sv in ds:
        if 2026 - sv.nam_sinh >= 20:
            print(sv.ma, sv.ten, sv.nam_sinh, sv.diem)

# =================== e. Đếm SV hệ DH ===================
def dem_he_DH(ds):
    dem = 0
    for sv in ds:
        if len(sv.ma) >= 4 and sv.ma[2:4] == "DH":
            dem += 1
    return dem

# =================== f. Đếm SV tên Lan ===================
def dem_ten_lan(ds):
    dem = 0
    for sv in ds:
        ten_cuoi = sv.ten.split()[-1]
        if ten_cuoi == "Lan":
            dem += 1
    return dem

# =================== g. Đếm SV họ Phan ===================
def dem_ho_phan(ds):
    dem = 0
    for sv in ds:
        ho = sv.ten.split()[0]
        if ho == "Phan":
            dem += 1
    return dem

# =================== MAIN ===================
ds = nhap_danh_sach()

print("\nSố SV lên lớp:", dem_len_lop(ds))
sv_20_tuoi(ds)
print("Số SV hệ DH:", dem_he_DH(ds))
print("Số SV tên Lan:", dem_ten_lan(ds))
print("Số SV họ Phan:", dem_ho_phan(ds))
```

```python
#1
def tron_hai_mang(a, b):
    ket_qua = []
    i = 0
    while i < len(a) and i < len(b):
        if a[i] < b[i]:
            ket_qua.append(a[i])
        else:
            ket_qua.append(b[i])
        i += 1

    if i < len(a):
        ket_qua.extend(a[i:])
    if i < len(b):
        ket_qua.extend(b[i:])

    return ket_qua

mang_a = [3, 9, 1, 4]
mang_b = [2, 7, 4, 3, 2, 8]
kq1 = tron_hai_mang(mang_a, mang_b)
print("Bài 1 - Mảng kết quả:", kq1)
```

```python
import numpy as np

# Định nghĩa các hàm xử lý ma trận
def giai_cau_2(a, n, x):
    print(f"--- Đang xử lý ma trận cấp {n}x{n} ---\n")

    # a. Tính tổng phần tử thuộc tam giác trên của đường chéo phụ (kể cả đường chéo phụ)
    # Quy luật: i + j <= n - 1
    tong_tg_phu = 0
    for i in range(n):
        for j in range(n - i):
            tong_tg_phu += a[i][j]
    print(f"a. Tổng tam giác trên chéo phụ: {tong_tg_phu}")

    # b. Chuyển các phần tử âm thành trị tuyệt đối
    for i in range(n):
        for j in range(n):
            if a[i][j] < 0:
                a[i][j] = abs(a[i][j])
    print(f"b. Ma trận sau khi lấy trị tuyệt đối: {a}")

    # c. Thay các phần tử chẵn bằng số nguyên x cho trước
    for i in range(n):
        for j in range(n):
            if a[i][j] % 2 == 0:
                a[i][j] = x
    print(f"c. Ma trận sau khi thay số chẵn bằng {x}: {a}")

    # d. Kiểm tra a có toàn chẵn không?
    toan_chan = True
    for i in range(n):
        for j in range(n):
            if a[i][j] % 2 != 0:
                toan_chan = False
                break
    print(f"d. Ma trận toàn chẵn: {toan_chan}")

    # e. Kiểm tra ma trận đối xứng (a[i][j] == a[j][i])
    doi_xung = True
    for i in range(n):
        for j in range(i + 1, n):
            if a[i][j] != a[j][i]:
                doi_xung = False
                break
    print(f"e. Ma trận đối xứng: {doi_xung}")

    # f. Kiểm tra đường chéo chính có tăng dần không
    cheo_chinh_tang = True
    for i in range(n - 1):
        if a[i][i] >= a[i+1][i+1]:
            cheo_chinh_tang = False
            break
    print(f"f. Đường chéo chính tăng dần: {cheo_chinh_tang}")

    # g. Xuất các phần tử thuộc tam giác dưới của đường chéo phụ (kể cả đường chéo phụ)
    # Quy luật: i + j >= n - 1
    print("g. Các phần tử thuộc tam giác dưới chéo phụ:")
    for i in range(n):
        for j in range(n - 1 - i, n):
            print(a[i][j], end=" ")
        print()

    # h. Kiểm tra đường chéo phụ có giảm dần không
    # Các phần tử chéo phụ: a[0][n-1], a[1][n-2], ..., a[n-1][0]
    cheo_phu_giam = True
    for i in range(n - 1):
        if a[i][n-1-i] <= a[i+1][n-1-(i+1)]:
            cheo_phu_giam = False
            break
    print(f"h. Đường chéo phụ giảm dần: {cheo_phu_giam}")

# --- CHƯƠNG TRÌNH CHÍNH ---

n_size = 3
x_value = 88
ma_tran_mau = [
    [1, 5, 3],
    [5, 4, 2],
    [3, 2, 6]
]

giai_cau_2(ma_tran_mau, n_size, x_value)
```
